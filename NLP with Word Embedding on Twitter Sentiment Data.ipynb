```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Embedding,Activation,Dropout
from tensorflow.keras.layers import Conv1D,MaxPooling1D,GlobalMaxPooling1D
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
from numpy import array
import pandas as pd
```

```python
df=pd.read_csv('data/sentiment140.csv',encoding='latin1',header=None)
```

```python
df
```

```python
df=df[[5,0]]
df.columns=['twits','sentiment']
df.head()
```

```python
df0=df[df['sentiment']==0].sample(2000)
df4=df[df['sentiment']==4].sample(2000)
```

```python
df=df0.append(df4).reset_index()
```

```python
df.head()
```

```python
df=df.drop(['index'],axis=1)
df
```

```python
text=df['twits'].tolist()
text
```

```python
y=df['sentiment']
```

```python
y=np.array(y)
```

```python
token=Tokenizer()
token.fit_on_texts(text)
```

```python
token
```

```python
token.word_index
```

```python
vocab_size=len(token.word_index)+1
vocab_size
```

```python
x=['i to the a my']
```

```python
token.texts_to_sequences(x)
```

```python
encoded_text=token.texts_to_sequences(text)
```

```python
print(encoded_text)
```

```python
max_len=0
for i in range(4000):    
    if len(encoded_text[i]) > max_len:
        max_len=len(encoded_text[i])
print(max_len)
```

```python
x=pad_sequences(encoded_text,maxlen=33,padding='post')
#just sequence of words, not vector
```

```python
print(x)
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2,stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape
```

```python
type(x_train),type(y_train)
```

```python
vec_size=300

model=Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=vec_size, input_length=33))

model.add(Conv1D(64,8,activation='relu'))
model.add(MaxPooling1D(2))
model.add(Dropout(0.2))

model.add(Dense(32,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(16,activation='relu'))

model.add(GlobalMaxPooling1D())
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
hist=model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
def get_encoded(x):
    x=token.texts_to_sequences(x)
    x=pad_sequences(x,maxlen=31,padding='post')
    return x
```
